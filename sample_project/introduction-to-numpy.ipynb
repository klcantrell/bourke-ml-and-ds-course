## Dot product

## Nut butter sales

```python
np.random.seed(88)
sales_amounts = np.random.randint(20, size=(5, 3))
sales_amounts
```

```python
# weekly sales dataframe
weekly_sales = pd.DataFrame(
    sales_amounts,
    index=["Mon", "Tues", "Wed", "Thurs", "Fri"],
    columns=["Almond butter", "Peanut butter", "Cashew butter"]
)
weekly_sales
```

```python
prices = np.array([10, 8, 2])
prices
```

```python
butter_prices = pd.DataFrame(prices.reshape(1, 3), columns=["Almond butter", "Peanut butter", "Cashew butter"], index=["Price"])
butter_prices
```

```python
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```

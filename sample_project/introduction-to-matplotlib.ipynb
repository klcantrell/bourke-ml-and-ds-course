```python
# one way to use object-oriented API
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# another way to use object-oriented API
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# third way to use object-oriented API (recommended)
fig, ax = plt.subplots()
ax.plot(x, y);
```

```python
nut_butter_prices = {"Almond butter": 10, "Peanut butter": 8, "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values())
plt.show()
```

```python
# histogram

x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

```python
# subplots
# option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(100))
plt.show()
```

```python
# subplots
# option 2

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(100))
plt.show()
```

```python
# from pandas dataframes

car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
# Clean Price column
car_sales["Price"] = car_sales["Price"].replace(r"[\$.,]", "", regex=True).astype(int) / 100

# Plot Price vs Odometer using OO API
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(car_sales["Odometer (KM)"], car_sales["Price"])

# Customize plot
ax.set(title="Price vs Odometer (OO API)",
       xlabel="Odometer (KM)",
       ylabel="Price ($)")

plt.show()
```

```python
# add fake dates
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
car_sales.plot(x="Sale Date", y="Total Sales")
plt.show()
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
# see distribution of age
heart_disease["age"].plot.hist(bins=50)
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10, 20), subplots=True)
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# pyplot method, limited
over_50.plot(kind="scatter", x="age", y="chol", c="target")
```

```python
# matplotlib oo method (sort of)

fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax) # actually, still mixed with pyplot method

```

```python
from matplotlib.lines import lineStyles
# matplotlib oo method, more control

fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50.age, y=over_50.chol, c=over_50.target)

ax.set(
    title="Heart Disease and Cholesterol Levels",
    xlabel="Age",
    ylabel="Cholesterol"
)

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(y=over_50.chol.mean(),linestyle="dashed")

plt.show()
```

```python
# subplots with cholesterol and thalach against age

fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# first sub plot
scatter = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target)
ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50.chol.mean(), linestyle="--")

# second sub plot
another_scatter = ax1.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target)
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")
ax1.legend(*another_scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50.thalach.mean(), linestyle="--")

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

plt.show()
```

```python

plt.style.use("seaborn-v0_8-whitegrid")

fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50.age, y=over_50.chol, c=over_50.target, cmap="winter")

ax.set(
    title="Heart Disease and Cholesterol Levels",
    xlabel="Age",
    ylabel="Cholesterol"
)

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(y=over_50.chol.mean(),linestyle="dashed")

plt.show()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# first sub plot
scatter = ax0.scatter(x=over_50.age, y=over_50.chol, c=over_50.target, cmap="winter")
ax0.set_xlim([50, 80])
ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50.chol.mean(), linestyle="--")

# second sub plot
another_scatter = ax1.scatter(x=over_50.age, y=over_50.thalach, c=over_50.target, cmap="summer")
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])
ax1.legend(*another_scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50.thalach.mean(), linestyle="--")

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

plt.show()
```
